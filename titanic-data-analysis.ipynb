# Summarize the dataset using descriptive statistics.


```python
print("the maximum age is:",titanic_data['Age'].max())
print("the minimum age is:",titanic_data['Age'].min())
print("the mean of the age is:",titanic_data['Age'].mean())
print("the mean of the age is:",titanic_data['Age'][titanic_data['Survived']==1].mean())
print("the median of the age is:",titanic_data['Age'].median())
print("the mode of the age is:",titanic_data['Age'].mode())
```

```python
titanic_data.fillna({"Age":titanic_data['Age'].mean()},inplace=True)
```

```python
titanic_data['Age']=titanic_data['Age'].astype(int)
```

```python
titanic_data['Age'].isna().sum()
```

```python
h_data=plt.hist(titanic_data['Age'],bins=30,color="green")
```

```python
titanic_data=titanic_data.drop('Cabin',axis=1)
```

```python
titanic_data.fillna({"Embarked":"S"},inplace=True)
```

```python
titanic_data
```

# Create a new feature FamilySize by combining the SibSp (siblings/spouses aboard) and Parch (parents/children aboard) columns.

```python
titanic_data['Family_size']=titanic_data['SibSp']+titanic_data['Parch']
```

# Gender: Compare survival rates between males and females.


```python
print("the number of servived males is :",len(titanic_data[(titanic_data['Survived'] == 1) & (titanic_data['Sex'] == "male")]))
print("the number of servived females is :",len(titanic_data[(titanic_data['Survived'] == 1) & (titanic_data['Sex'] == "female")]))
print("the percentage of servived females is :",len(titanic_data[(titanic_data['Survived'] == 1) & (titanic_data['Sex'] == "male")])/342*100)
print("the percentage of servived females is :",len(titanic_data[(titanic_data['Survived'] == 1) & (titanic_data['Sex'] == "female")])/342*100)
```

# Pclass: Compare survival rates across different passenger classes.


```python
print("the number of survived people in class 1 is:",len(titanic_data[(titanic_data['Pclass']==1) & (titanic_data['Survived']==1)]))
print("the number of survived people in class 2 is:",len(titanic_data[(titanic_data['Pclass']==2) & (titanic_data['Survived']==1)]))
print("the number of survived people in class 3 is:",len(titanic_data[(titanic_data['Pclass']==3) & (titanic_data['Survived']==1)]))
print("the percentage of survived people in class 1 is:",len(titanic_data[(titanic_data['Pclass']==1) & (titanic_data['Survived']==1)])/342*100)
print("the percentage of survived people in class 2 is:",len(titanic_data[(titanic_data['Pclass']==2) & (titanic_data['Survived']==1)])/342*100)
print("the percentage of survived people in class 3 is:",len(titanic_data[(titanic_data['Pclass']==3) & (titanic_data['Survived']==1)])/342*100)
```

# Embarked: Analyze the survival rate based on the port of embarkation.

# Use Plotly to compare survival rates across different categories (e.g., gender, Pclass, Embarked). Make the charts interactive to allow for zooming and hovering over data points.

```python
import plotly.express as px
px.bar(titanic_data,x='Sex',y='Survived',title="Survival Rates based on Sex",labels={'Survived':"survival rate"},color="Sex",text="Sex")
```

```python
fig1=px.bar(titanic_data,x='Pclass',y='Survived',title="Survival Rates based on Class",labels={'Survived':"survival rate"},color="Pclass",text="Pclass")
fig2=px.bar(titanic_data,x='Pclass',y='Survived',title="Survival Rates based on Class and Sex",labels={'Survived':"survival rate"},color="Sex",text="Pclass")
fig1.show()
fig2.show()
```

```python
fig3=px.bar(titanic_data,x='Embarked',y='Survived',title="Survival Rates based on Embarked",labels={'Survived':"survival rate"},color="Embarked",text="Embarked")
fig4=px.bar(titanic_data,x='Embarked',y='Survived',title="Survival Rates based on Embarked and Sex",labels={'Survived':"survival rate"},color="Sex",text="Embarked")
fig3.show()
fig4.show()
```

# Show the distribution of age and fare for survivors vs. non-survivors using Plotly, allowing for a more dynamic exploration of the data.
